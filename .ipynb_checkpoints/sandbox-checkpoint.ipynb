```python
def get_ticker_info(ticker):

    stock = yf.Ticker(ticker)

    company_name = stock.info['shortName']
    sector = stock.info['sector']
    currency = stock.info['currency']
    exchange = stock.info['exchange']
    
    ticker_info = pd.DataFrame({'ticker': [ticker], 'name': [company_name], 'sector': [sector], 'currency': [currency], 'exchange': [exchange]})
    
    return ticker_info

def get_ticker_prices(ticker, start_dt, end_dt):
    
    prices = stocks.get_data(ticker, start_date = start_dt, end_date = end_dt)
    
    prices.index.name = 'date'
    prices.reset_index(inplace=True)
    
    return prices
    
```

```python

ticker_prices = get_ticker_prices('BA','2020-01-01','2020-04-30')
ticker_prices
```

```python
import psycopg2
from modules.config import config

def insert_ticker(ticker_info, time_stamp):
    """ insert a ticker into the symbol """
    sql = """
    INSERT into stocks.symbol(ticker, name, sector, currency, created_date, last_updated_date)
    values('%s','%s','%s','%s','%s','%s');
    """ %(
        ticker_info['ticker'].values[0], 
        ticker_info['name'].values[0], 
        ticker_info['sector'].values[0], 
        ticker_info['currency'].values[0],
        time_stamp,
        time_stamp
    )
    conn = None
    vendor_id = None
    try:
        # read database configuration
        params = config()
        # connect to the PostgreSQL database
        conn = psycopg2.connect(**params)
        # create a new cursor
        cur = conn.cursor()
        # execute the INSERT statement
        cur.execute(sql, ticker_info)
        # commit the changes to the database
        conn.commit()
        # close communication with the database
        cur.close()
    except (Exception, psycopg2.DatabaseError) as error:
        print(error)
    finally:
        if conn is not None:
            conn.close()

    return vendor_id
```

```python
from datetime import datetime

now = datetime.now()

time_stamp = now.strftime("%Y-%m-%d %H:%M")

ticker_info = get_ticker_info('MSFT')
insert_ticker(ticker_info, time_stamp)
```

```python
from datetime import datetime

now = datetime.now()

time_stamp = now.strftime("%Y-%m-%d %H:%M")

sp = stocks.tickers_sp500()

for s in sp:
    try:
        ticker_info = get_ticker_info(s)
        insert_ticker(ticker_info, time_stamp)
    except: 
        print(s)
```
